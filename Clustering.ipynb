```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv

class Net(torch.nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = GCNConv(dataset.num_node_features, 16)
        self.conv2 = GCNConv(16, dataset.num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index

        x = self.conv1(x, edge_index)
        h = x.clone()
        x = F.relu(x)
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, edge_index)

        return F.log_softmax(x, dim=1), h
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = Net().to(device)
data = dataset[0].to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)

model.train()
for epoch in range(200):
    optimizer.zero_grad()
    out, h = model(data)
    #print(out.shape)
    #print(h.shape)
    loss = F.nll_loss(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
```

```python
model.eval()
out, hidden = model(data)
_, pred = out.max(dim=1)
correct = float (pred[data.test_mask].eq(data.y[data.test_mask]).sum().item())
acc = correct / data.test_mask.sum().item()
print('Accuracy: {:.4f}'.format(acc))
```

```python
from torch_geometric.data import Data

def create_ladder(n):
    edges = [(0, 1), (1, 0)]

    for i in range(2, 2 * n, 2):
        edges.append((i, i + 1))
        edges.append((i + 1, i))
        #nx_edges.append((i, i + 1))
        edges.append((i, i - 2))
        edges.append((i - 2, i))
        #nx_edges.append((i, i - 2))
        edges.append((i + 1, i - 1))
        edges.append((i - 1, i + 1))
        #nx_edges.append((i + 1, i - 1))

    edge_index = torch.tensor(edges, dtype=torch.long)
    #self.nx_graph.add_edges_from(nx_edges)
    x = torch.zeros((2*n, 1433)) + 0.5
    y = torch.randint(0, 7, (2*n,))
    edge_index = edge_index.t().contiguous()
    graph = Data(x=x, edge_index=edge_index, y=y)
    return graph
```

```python
def create_grid(n, m):
    edges = []

    for j in range(m - 1):
        edges.append((j, j + 1))
        edges.append((j + 1, j))
        #nx_edges.append((j, j + 1))

    for i in range(1, n):
        for j in range(m - 1):
            edges.append((m*i + j, m*i + j + 1))
            edges.append((m*i + j + 1, m*i + j))
            #nx_edges.append((m*i + j, m*i + j + 1))
            edges.append((m*(i-1) + j, m*i + j))
            edges.append((m*i + j, m*(i-1) + j))
            #nx_edges.append((m*(i-1) + j, m*i + j))
        edges.append((m*(i-1) + m - 1, m*i + m - 1))
        edges.append((m*i + m - 1, m*(i-1) + m - 1))
        #nx_edges.append((m*(i-1) + m - 1, m*i + m - 1))

    edge_index = torch.tensor(edges, dtype=torch.long)
    x = torch.zeros((n*m, 1433)) + 0.5
    y = torch.randint(0, 7, (m*n,))
    edge_index = edge_index.t().contiguous()
    graph = Data(x=x, edge_index=edge_index, y=y)
    return graph
```

```python
ladder_1 = create_ladder(5)
ladder_2 = create_ladder(6)
ladder_3 = create_ladder(1)
grid_1 = create_grid(2, 3)
```

```python
_, hidden_1 = model(ladder_1)
_, hidden_2 = model(ladder_2)
_, hidden_3 = model(ladder_3)
_, hidden_4 = model(grid_1)
```
